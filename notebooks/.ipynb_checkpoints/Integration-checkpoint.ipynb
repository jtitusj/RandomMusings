```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
n = 10

def get_rectangles(xmin, xmax, n):
    dx = (xmax - xmin)/n
    dxs = np.arange(n)*dx
    return dxs
```

```python
xmin = 0
xmax = 1
x = np.linspace(xmin, xmax, 100000)
f = lambda x: 5*x**5 - 4*x**4 + 3*x**3 - 2*x**2 + x + 1

s = 10
fig, ax = plt.subplots(s, figsize=(6, 40), sharex=True)

for i in range(s):
    ax[i].plot(x, f(x), color='r')
    if i > 0:
        dxs = get_rectangles(xmin, xmax, 10*i)
        ax[i].bar(x=dxs, height=f(dxs), edgecolor='k', align='edge',
               label=f'sum_of_bars={(f(dxs)*dxs[1]).sum():.4f}')
        ax[i].legend()
    ax[i].set_xlim(x[0], x[-1])
    ax[i].set_ylim(0,)
    
fig.tight_layout()
```
